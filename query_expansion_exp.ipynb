```python
from pypdf import PdfReader
import re
```

```python
# helper function to read pdf document
def read_pdf_content(pdf_path):
    full_text = ''
    # Open the PDF file in read-binary mode
    with open(pdf_path, 'rb') as file:
        reader = PdfReader(file)
        
        full_text = ""
        # Loop through all the pages and extract text
        for page in reader.pages:
            full_text += page.extract_text() + "\n"
            
    return full_text
```

```python
# Helper function to clean the text, removing page numbers, extra spaces, etc.
def clean_text(text):

    # remove page numbers
    text = re.sub(r'\n\d+\n', '\n', text)

    # remove extra spaces
    text = re.sub(r'\s+', ' ', text)

    # restore line breaks before section numbers
    text = re.sub(r'(\d{1,2}\.)', r'\n\1', text)

    return text.strip()
```

```python
import re
from nltk.corpus import stopwords

# custom stopwords (safe for legal text)
base_stopwords = set(stopwords.words("english"))

# remove only generic ones
custom_stopwords = base_stopwords - {
    "not", "no", "nor", "shall", "must", "may"
}

def clean_text_v2(text):
    # 1. remove page artifacts
    text = re.sub(r"<PARSED TEXT FOR PAGE:.*?>", " ", text)

    # 2. lowercase
    text = text.lower()

    # 3. remove punctuation (keep numbers)
    text = re.sub(r"[^a-z0-9\s]", " ", text)

    # 4. tokenize
    tokens = text.split()

    # 5. remove stopwords
    tokens = [t for t in tokens if t not in custom_stopwords]

    # 6. remove short tokens
    tokens = [t for t in tokens if len(t) > 2]

    return " ".join(tokens)
```

```python
# helper function to split the sections from given document: Section wise chunking
def split_sections(text):

    sections = []
    current_section = None
    expected_section = 1

    lines = text.split("\n")

    for line in lines:

        line = line.strip()
        if not line:
            continue

        match = re.match(r'^(\d{1,2})\.?\s+(.+)', line)

        if match:

            section_number = int(match.group(1))

            if section_number == expected_section:

                if current_section:
                    sections.append(current_section)

                current_section = {
                    "header": f"{section_number}. {match.group(2)}",
                    "content": line + "\n"
                }

                expected_section += 1
                continue

        if current_section:
            current_section["content"] += line + "\n"

    if current_section:
        sections.append(current_section)

    return sections
```

```python
# helper function to do fixed size chunking
def chunk_sections(sections, chunk_size=400):

    chunks = []
    chunk_id = 0

    for sec in sections:

        words = sec["content"].split()

        for i in range(0, len(words), chunk_size):

            chunk_text = " ".join(words[i:i+chunk_size])

            chunks.append({
                "chunk_id": chunk_id,
                "section_title": sec["header"],
                "text": chunk_text,
                "source": "axis_credit_card_doc"
            })

            chunk_id += 1

    return chunks
```

```python
def add_bigrams(tokens):
    bigrams = []
    for i in range(len(tokens) - 1):
        bigrams.append(tokens[i] + "_" + tokens[i+1])
    return tokens + bigrams
```

```python
# Helper function to get the similar chunks using BM25
def bm25_retrieve(query, bm25, chunks, top_k=5):

    tokenized_query = query.lower().split()

    scores = bm25.get_scores(tokenized_query)

    top_indices = np.argsort(scores)[::-1][:top_k]

    # results = [chunks[i] for i in top_indices]

    return top_indices
```

```python
result_bm25 = []
for q in evaluation_queries:
    query = q['query']
    query = clean_text_v2(query)
    relevant_chunks = q['relevant_chunks']
    
    indices = bm25_retrieve(query, bm25, final_chunks, top_k=10)

    result_bm25.append({
        "query": query,
        "actual": relevant_chunks,
        "retrieved": indices
    })
```

#### BM25 Baseline validation

```python
import numpy as np

precision_scores = []
recall_scores = []
mrr_scores = []

K = 5

for item in result_bm25:

    retrieved = item["retrieved"].tolist()[:5]
    relevant = item["actual"]

    # Count how many retrieved are relevant
    relevant_retrieved = len(set(retrieved) & set(relevant))

    # Precision@K
    precision = relevant_retrieved / K
    precision_scores.append(precision)

    # Recall@K
    recall = relevant_retrieved / len(relevant)
    recall_scores.append(recall)

    # MRR (first relevant chunk rank)
    rank = None
    for i, r in enumerate(retrieved):
        if r in relevant:
            rank = i + 1
            break

    if rank:
        mrr_scores.append(1 / rank)
    else:
        mrr_scores.append(0)

precision = np.mean(precision_scores)
recall = np.mean(recall_scores)
mrr = np.mean(mrr_scores)

print("Precision@5:", precision)
print("Recall@5:", recall)
print("MRR:", mrr)
```

```python
QUERY_EXPANSION = {
    # Fees / Charges
    "fee": ["charge", "penalty", "cost"],
    "fees": ["charges", "penalties", "costs"],
    "late payment": ["late fee", "penalty", "overdue charge"],
    "finance charges": ["interest", "borrowing cost"],
    
    # Credit Limit
    "credit limit": ["maximum credit", "limit", "spending limit"],
    "increase": ["enhance", "raise", "extend"],
    "reduce": ["decrease", "lower"],
    
    # Cardmember / User
    "cardmember": ["cardholder", "customer", "user"],
    "primary cardmember": ["main cardholder"],
    "additional card": ["add-on card", "supplementary card"],
    
    # Transactions
    "transaction": ["purchase", "payment", "usage"],
    "domestic": ["india"],
    "international": ["overseas", "foreign"],
    
    # Billing / Payment
    "billing cycle": ["statement period"],
    "statement": ["billing statement"],
    "minimum amount due": ["minimum payment", "due amount"],
    "payment due date": ["due date"],
    
    # Cash Advance
    "cash advance": ["cash withdrawal", "atm withdrawal"],
    "atm": ["cash machine"],
    
    # Lost / Stolen
    "lost": ["stolen", "missing"],
    "stolen": ["lost", "unauthorized use"],
    "liability": ["responsibility"],
    
    # Insurance
    "insurance": ["coverage", "protection"],
    "accident": ["injury", "death cover"],
    
    # Termination / Cancellation
    "terminate": ["cancel", "close"],
    "cancel": ["terminate", "close"],
    
    # Default / Legal
    "default": ["non payment", "failure to pay"],
    "legal": ["law", "regulation"],
    "arbitration": ["dispute resolution"],
    
    # Bank Actions
    "bank": ["issuer", "lender"],
    
    # General
    "charges": ["fees", "costs"],
    "responsibility": ["liability", "obligation"],
    "rules": ["conditions", "terms"]
}
```

```python
def expand_query(query):
    query = query.lower()
    expanded_terms = set([query])

    for key, values in QUERY_EXPANSION.items():
        if key in query:
            expanded_terms.update(values)

    return query + " " + " ".join(expanded_terms)
```
